```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = "D:/UMD/Project/dataset_10/Data-setvIPL2019.4(sold_players).csv"
df = pd.read_csv(file_path)
df.head()
```

```python
import re
df["price_sold"] = df["price_sold"].apply(lambda x : x.replace(",",""))
df["base_price"] = df["base_price"].apply(lambda x : x.replace(",",""))
```

```python
df.head(2)
```

```python
df = df.astype({'base_price':np.int64})
df = df.astype({'price_sold':np.int64})
```

```python
df.info()
```

```python
def category_encoder(y):
    if str(y) == "Batsman":
        y = 1
    elif str(y) == "Bowler":
        y = 2
    elif str(y) == "Wicket Keeper":
        y = 3
    else:
        y = 4
    return y

df["category"] = df["category"].apply(lambda x : category_encoder(x))
```

```python
X = df.drop(columns =["player_name", "price_sold"]).values
y = df["price_sold"].values

y = y.reshape(-1,1)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 28)

y_true = y_test
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train, y_train)

y_predict = reg.predict(X_test)
```

```python
coef = reg.coef_
intercept = reg.intercept_
print("coefficients ", coef)
print("intercept ", intercept)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from math import sqrt
RMSE = sqrt(mean_squared_error(y_true, y_predict))
RMSE
```

```python
MAE = mean_absolute_error(y_true, y_predict)
MAE
```
